```python
def f(x):
    return x*x

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h = 1e-10

def ff(x):
    return (f(x+h)-f(x))/h

def fb(x):
    return (f(x)-f(x-h))/h

def fc(x):
    return (f(x+h)-f(x-h))/2*h

x = np.arange(1,100,1,dtype=float)
yff, yfb, yfc = np.zeros(len(x)), np.zeros(len(x)), np.zeros(len(x))
#plt.hold(True)
plt.plot(x,ff(x), 'o', fb(x), '-', fc(x)*1e+20, '.')
#plt.show()
```

```python
def f(x):
    return x*x*x

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#h = 1e-6

def f23(x):
    return (f(x+h)+f(x-h)-2.0*f(x))/(h*h)

def f25(x):
    return (f(x-2.0*h)-8.0*f(x-h)+8.0*f(x+h)-f(x+2.0*h))/(12.0*h)

x = np.arange(1,100,1,dtype=float)
plt.hold(True)
for i in xrange(1,10):
    h = 1.0/np.power(10,i)
    plt.ylim(-10,600)
    plt.plot(x, f23(x), '-', f25(x), 'o')

plt.show()
```
