```python
def f(x):
    return x*x

def ff(x):
    return (f(x+h)-f(x))/h

def fb(x):
    return (f(x)-f(x-h))/h

def fc(x):
    return (f(x+h)-f(x-h))/(2.0*h)

def f5(x):
    return (f(x-2.0*h)-8.0*f(x-h)+8.0*f(x+h)-f(x+2.0*h))/(12.0*h)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h = 1e-7
x = np.arange(1,100,1,dtype=float)

plt.hold(True)
plt.plot(x, abs(ff(x) - 2.0*x), '^', label = 'forward difference')
plt.plot(x, abs(fb(x)- 2.0*x), 'v', label = 'backward difference')
plt.plot(x, abs(fc(x)- 2.0*x), '+', label = 'central difference')
plt.plot(x, abs(f5(x)- 2.0*x),'*', label = '5 point difference')
plt.legend(loc='upper left')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return x*x*x

def f23(x):
    return (f(x+h)+f(x-h)-2.0*f(x))/(h*h)

x = np.arange(1,100,1,dtype=float)
plt.hold(True)
for i in xrange(1,10):
    h = 1.0/np.power(10,i)
    plt.ylim(-50,50)
    plt.plot(x, abs(f23(x)-6.0*x), label = i)

plt.legend(loc='upper left')
plt.show()
```
