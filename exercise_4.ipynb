```python
def f(a, x):
    return np.exp(a*x)

def ff(x):
    return (f(a, x+h)-f(a, x))/h

def fb(x):
    return (f(a, x)-f(a, x-h))/h

def fc(x):
    return (f(a, x+h)-f(a, x-h))/(2.0*h)

def f5(x):
    return (f(a, x-2.0*h)-8.0*f(a, x-h)+8.0*f(a, x+h)-f(a, x+2.0*h))/(12.0*h)

#a = input()

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

h = 1e-8
x = np.arange(1,25,1,dtype=float)
for a in xrange(1,2):
    plt.hold(True)
    #plt.ylim(0,1000)
    plt.plot(x, abs(ff(x) - a*f(a,x)), '^', label = 'forward difference')
    plt.plot(x, abs(fb(x)- a*f(a,x)), 'v', label = 'backward difference')
    plt.plot(x, abs(fc(x)- a*f(a,x)), '+', label = 'central difference')
    plt.plot(x, abs(f5(x)- a*f(a,x)),'*', label = '5 point difference')
    plt.legend(loc='upper left')    
```
